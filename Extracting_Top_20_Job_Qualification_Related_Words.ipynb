# TfidfVectorizer with text preprocessing (Stopwords+  Stemming+ removing punctuations & numbers)

```python
string.punctuation
```

```python
#remove punctuation & numbers in each sentence

def remove_punc_func(x):
    new=''
    for a in x:
        if a not in string.punctuation:
            new=new+a
    return new



def remove_number_func (x): 
    new1=''
    for a in x:
        if a.isdigit()==False:
            new1=new1+a
    return new1



def compose(f, g):
    return lambda x: f(g(x))


function=compose(remove_number_func,remove_punc_func)

#remove punctuation & numbers in the overall text 
total_removed_punc_num=[]
for a in total:
    total_removed_punc_num.append(function(a))
total_removed_punc_num
```

```python
def tokenize_lemma(text):
    return [w.lemma_.lower() for w in nlp(text)]

stop_words_lemma = set(tokenize_lemma(' '.join(STOP_WORDS.union({'telstra','experience'}))))       #use union
```

```python
tfidfv=TfidfVectorizer(stop_words=stop_words_lemma,
                       tokenizer=tokenize_lemma, 
                       token_pattern=None )

tfidfv.fit_transform(total_removed_punc_num)

new=[]
for x in range(0, 382):
    new.append(tfidfv.get_feature_names()[x])
```

```python
df=pd.DataFrame(tfidfv.fit_transform(total_removed_punc_num).todense(), columns=new)
df
```

```python
final=[]
for x in new:
    final.append((x, df[x].sum()))
    
sorted(final, key=lambda x: -x[1]);
```

```python
top_20_qualification_related_pairs=sorted(final, key=lambda x: -x[1])[1:21] 
top_20_qualification_related_words= [x[0] for x in top_20_qualification_related_pairs]
top_20_qualification_related_words
```
